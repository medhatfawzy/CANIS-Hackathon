```python
def plot_count(df, col, min_count=5, figsize=(8, 6)):
    # Assuming canis_data is your DataFrame and "Region of Focus" is the column of interest
    counts = df[col].value_counts()
    filtered_counts = counts[counts > min_count]

    # Create horizontal bar plot
    ax = filtered_counts.plot(kind="barh", color="blue", figsize=figsize)

    # Add the text annotations on the bars
    for index, value in enumerate(filtered_counts):
        # Set a margin for the text (e.g., 5% of the max value)
        margin = max(filtered_counts) * 0.05
        # Position the text to the right of the bar
        text_position = value + margin if value > 10 else value + margin / 2

        ax.text(text_position, index, f' {value}', va='center', ha='right')

    plt.title(col)
    # Display the plot
    plt.show()
```

```python
plot_count(canis_data, "Region of Focus")
```

```python
plot_count(canis_data, "Language")
```

```python
plot_count(canis_data, "Entity owner (English)")
```

```python
plot_count(canis_data, "Parent entity (English)")
```

```python
opts.defaults(opts.Graph(width=400, height=400))

# Declare abstract edges
N = 8
source = np.zeros(N)
target = np.arange(N)

simple_graph = hv.Graph(((source, target),))
simple_graph
```

```python
simple_graph.opts(directed=True, arrowhead_length=0.05)
```

```python
simple_graph.edgepaths + simple_graph.nodes
```

```python
# Node info
np.random.seed(7)
x, y = simple_graph.nodes.array([0, 1]).T
node_labels = ['Output']+['Input']*(N-1)
edge_weights = np.random.rand(8)

# Compute edge paths
def bezier(start, end, control, steps=np.linspace(0, 1, 100)):
    return (1-steps)**2*start + 2*(1-steps)*steps*control+steps**2*end

paths = []
for node_index in target:
    ex, ey = x[node_index], y[node_index]
    paths.append(np.column_stack([bezier(x[0], ex, 0), bezier(y[0], ey, 0)]))

# Declare Graph
nodes = hv.Nodes((x, y, target, node_labels), vdims='Type')
graph = hv.Graph(((source, target, edge_weights), nodes, paths), vdims='Weight')

graph.opts(node_color='Type', edge_color='Weight', cmap=['blue', 'red'], edge_cmap='viridis')
```
