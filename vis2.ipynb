```python
continent_dict = {
    "Europe": [
        "Anglosphere",
        "Europe",
        "la Francophonie",
        "Russia",
        "Hispanophone",
        "Lusophone",
    ],
    "Asia": [
        "China",
        "Middle East and North Africa",
        "Southeast Asia",
        "East Asia",
        "South Asia",
        "Central Asia",
    ],
    "Africa": [
        "African Union",
        "Southern Africa",
        "East Africa",
        "Central Africa",
        "West Africa",
    ],
    "North America": [
        "North America",
        "Caribbean",
        "Dominican Republic",
        "Central America",
    ],
    "South America": ["South America"],
    "Worldwide": ["United Nations", "UN"],
    "Oceania": ["Oceania"],
}

df["Continent of Focus"] = df["Region of Focus"].apply(
    lambda x: next(
        (continent for continent, regions in continent_dict.items() if x in regions),
        "Unknown",
    )
)
```

```python
twitter_df = df[
    [
        "Name (English)",
        "Region of Focus",
        "Continent of Focus",
        "Entity owner (English)",
        "Parent entity (English)",
        "X (Twitter) handle",
        "X (Twitter) URL",
        "X (Twitter) Follower #",
    ]
].dropna()

facebook_df = df[
    [
        "Name (English)",
        "Region of Focus",
        "Continent of Focus",
        "Entity owner (English)",
        "Parent entity (English)",
        "Facebook page",
        "Facebook URL",
        "Facebook Follower #",
    ]
].dropna()

instagram_df = df[
    [
        "Name (English)",
        "Region of Focus",
        "Continent of Focus",
        "Entity owner (English)",
        "Parent entity (English)",
        "Instragram page",
        "Instagram URL",
        "Instagram Follower #",
    ]
].dropna()

threads_df = df[
    [
        "Name (English)",
        "Region of Focus",
        "Continent of Focus",
        "Entity owner (English)",
        "Parent entity (English)",
        "Threads account",
        "Threads URL",
        "Threads Follower #",
    ]
].dropna()


youtube_df = df[
    [
        "Name (English)",
        "Region of Focus",
        "Continent of Focus",
        "Entity owner (English)",
        "Parent entity (English)",
        "YouTube account",
        "YouTube URL",
        "YouTube Subscriber #",
    ]
].dropna()

tiktok_df = df[
    [
        "Name (English)",
        "Region of Focus",
        "Continent of Focus",
        "Entity owner (English)",
        "Parent entity (English)",
        "TikTok account",
        "TikTok URL",
        "TikTok Subscriber #",
    ]
].dropna()

dfs = {
    "twitter": twitter_df,
    "facebook": facebook_df,
    "instagram": instagram_df,
    "threads": threads_df,
    "youtube": youtube_df,
    "tiktok": tiktok_df,
}

for _df in dfs:
    dfs[_df].columns = [
        "Name",
        "Region",
        "Continent",
        "Entity",
        "Parent",
        "Account",
        "URL",
        "Followers",
    ]
```

# Basic Visualization

```python
fig, ax = plt.subplots(2, 3, figsize=(20, 10))
ax = ax.flatten()
supertitle = fig.suptitle(
    "State Media on Social Media Platforms (Entity per Region)",
    fontsize=20,
    fontweight="bold",
)
for i, _df in enumerate(dfs):
    sns.scatterplot(
        data=dfs[_df].sort_values("Followers", ascending=False),
        x="Entity",
        y="Region",
        size="Followers",
        hue="Region",
        ax=ax[i],
    )
    ax[i].set_title(_df, fontsize=16, fontweight="bold")
    ax[i].set_xlabel("Entity", fontsize=12)
    ax[i].set_ylabel("Region", fontsize=12)
    ax[i].legend().remove()
    ax[i].grid(False)
    ax[i].xaxis.set_visible(False)

for ax in fig.axes:
    plt.sca(ax)
    plt.xticks(rotation=60, ha="right")

plt.tight_layout()
plt.show()
```
