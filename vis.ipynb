```python
def plot_count(df, col, min_count=5, figsize=(8, 6)):
    # Assuming canis_data is your DataFrame and "Region of Focus" is the column of interest
    counts = df[col].value_counts()
    filtered_counts = counts[counts > min_count]

    # Create horizontal bar plot
    ax = filtered_counts.plot(kind="barh", color="blue", figsize=figsize)

    # Add the text annotations on the bars
    for index, value in enumerate(filtered_counts):
        # Set a margin for the text (e.g., 5% of the max value)
        margin = max(filtered_counts) * 0.05
        # Position the text to the right of the bar
        text_position = value + margin if value > 10 else value + margin / 2

        ax.text(text_position, index, f' {value}', va='center', ha='right')

    plt.title(col)
    # Display the plot
    plt.show()
```

```python
plot_count(canis_data, "Region of Focus")
```

```python
plot_count(canis_data, "Language")
```
