```python
twitter_df = df[
    [
        "Name (English)",
        "Region of Focus",
        "Entity owner (English)",
        "Parent entity (English)",
        "X (Twitter) handle",
        "X (Twitter) URL",
        "X (Twitter) Follower #",
    ]
].dropna()

facebook_df = df[
    [
        "Name (English)",
        "Region of Focus",
        "Entity owner (English)",
        "Parent entity (English)",
        "Facebook page",
        "Facebook URL",
        "Facebook Follower #",
    ]
].dropna()

instagram_df = df[
    [
        "Name (English)",
        "Region of Focus",
        "Entity owner (English)",
        "Parent entity (English)",
        "Instragram page",
        "Instagram URL",
        "Instagram Follower #",
    ]
].dropna()

threads_df = df[
    [
        "Name (English)",
        "Region of Focus",
        "Entity owner (English)",
        "Parent entity (English)",
        "Threads account",
        "Threads URL",
        "Threads Follower #",
    ]
].dropna()


youtube_df = df[
    [
        "Name (English)",
        "Region of Focus",
        "Entity owner (English)",
        "Parent entity (English)",
        "YouTube account",
        "YouTube URL",
        "YouTube Subscriber #",
    ]
].dropna()

tiktok_df = df[
    [
        "Name (English)",
        "Region of Focus",
        "Entity owner (English)",
        "Parent entity (English)",
        "TikTok account",
        "TikTok URL",
        "TikTok Subscriber #",
    ]
].dropna()

dfs = {
    "twitter": twitter_df,
    "facebook": facebook_df,
    "instagram": instagram_df,
    "threads": threads_df,
    "youtube": youtube_df,
    "tiktok": tiktok_df,
}

for _df in dfs:
    dfs[_df].columns = [
        "Name",
        "Region",
        "Entity",
        "Parent",
        "Account",
        "URL",
        "Followers",
    ]
    dfs[_df]["Category"] = _df
```

# Basic Visualization

```python
fig, ax = plt.subplots(2, 3, figsize=(20, 10))
ax = ax.flatten()
supertitle = fig.suptitle(
    "State Media on Social Media Platforms (Entity per Region)",
    fontsize=20,
    fontweight="bold",
)
for i, _df in enumerate(dfs):
    sns.scatterplot(
        data=dfs[_df].sort_values("Followers", ascending=False),
        x="Entity",
        y="Region",
        size="Followers",
        hue="Region",
        ax=ax[i],
    )
    ax[i].set_title(_df, fontsize=16, fontweight="bold")
    ax[i].set_xlabel("Entity", fontsize=12)
    ax[i].set_ylabel("Region", fontsize=12)
    ax[i].legend().remove()
    ax[i].grid(False)
    ax[i].xaxis.set_visible(False)

for ax in fig.axes:
    plt.sca(ax)
    plt.xticks(rotation=60, ha="right")

plt.tight_layout()
plt.show()
```

```python
class_coordinates = {
    "Anglosphere": (51.5, -0.12),
    "la Francophonie": (48.9, 2.4),
    "Lusophone": (38.7, -9.1),
    "Hispanophone": (40.4, -3.7),
    "China": (39.9, 116.4),
    "Russia": (55.8, 37.6),
    "Europe": (52.5, 5.5),
    "Middle East and North Africa": (30.0, 30.0),
    "Southeast Asia": (1.3, 103.8),
    "East Asia": (35.7, 139.7),
    "South Asia": (28.6, 77.2),
    "Central Asia": (41.3, 69.3),
    "East Africa": (-6.2, 35.7),
    "Central Africa": (0.5, 25.5),
    "Southern Africa": (-25.7, 28.2),
    "West Africa": (6.3, -1.2),
    "Caribbean": (18.1, -77.3),
    "Oceania": (-33.9, 151.2),
    "North America": (38.9, -77.0),
    "Central America": (9.9, -84.1),
    "South America": (-23.5, -46.6),
    "United Nations": (40.7, -74.0),
}
```

```python
gdf = pd.DataFrame(columns=["Region", "Followers", "Category", "Latitude", "Longitude"])

for _df in dfs:
    for index, row in dfs[_df].iterrows():
        coordinates = class_coordinates[row["Region"]]
        gdf = pd.concat(
            [
                gdf,
                pd.DataFrame(
                    {
                        "Region": [row["Region"]],
                        "Followers": [row["Followers"]],
                        "Category": [row["Category"]],
                        "Latitude": [coordinates[0]],
                        "Longitude": [coordinates[1]],
                    }
                ),
            ],
            ignore_index=True,
        )
```

```python
gdf = gdf.groupby(["Region", "Category"]).sum().reset_index()
```
